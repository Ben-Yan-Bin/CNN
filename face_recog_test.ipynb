```python
import face_recognition
import cv2
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
```

```python
# filename = "test.jpg"
# filename = "img/img_17399.jpg"
# filename = "img/img_81170.jpg"
# filename = "img/img_17406.jpg"
# filename = "img/img_81155.jpg"
# filename1 = "obama3.jpg"
# filename2 = "obama.jpg"
filename1 = "origin/YB1.jpg"
filename2 = "origin/YB3.jpg"

# images = glob.glob("img/*.jpg")

image1 = face_recognition.load_image_file(filename1)
image2 = face_recognition.load_image_file(filename2)
img1 = mpimg.imread(filename1)
imgcp1 = np.copy(img1)
img2 = mpimg.imread(filename2)
imgcp2 = np.copy(img2)
```

```python
face_locations1 = face_recognition.face_locations(image1)
face_locations2 = face_recognition.face_locations(image2)
image1_encoding = face_recognition.face_encodings(image1)[0]
image2_encoding = face_recognition.face_encodings(image2)[0]
# print(img.shape)
print(face_locations1)
print(face_locations2)
```

```python
results = face_recognition.compare_faces([image1_encoding], image2_encoding, tolerance=0.4)
diff = face_recognition.face_distance([image1_encoding], image2_encoding)
if results[0]:
    print("Same person!")
else:
    print("Different person!")

print("Difference = ", diff)
```

```python
for i in range(1):
    x10 = face_locations1[i][3]
    y10 = face_locations1[i][0]
    x11 = face_locations1[i][1]
    y11 = face_locations1[i][2]
    x20 = face_locations2[i][3]
    y20 = face_locations2[i][0]
    x21 = face_locations2[i][1]
    y21 = face_locations2[i][2]
    imgcp1 = cv2.rectangle(imgcp1, (x10, y10), (x11, y11), (0,255,0), 5)
    imgcp2 = cv2.rectangle(imgcp2, (x20, y20), (x21, y21), (0,255,0), 5)
```

```python
fig = plt.figure(figsize = (20, 20))

plt.subplot(221)
plt.imshow(img1)
plt.subplot(222)
plt.imshow(imgcp1)

plt.subplot(223)
plt.imshow(img2)
plt.subplot(224)
plt.imshow(imgcp2)

plt.show()
```
